# Funkcje

Jak zdefiniować ten kawałek? \
Otóż w taki sposób robimy przepis na funkcję:
```
def nazwa_funkcji():
    print("Moja funkcja")
    print("Tutaj funkcja coś robi")
```


### A gdybyśmy chcieli, żeby jedną linijką robić kreski na początku, potem właściwy tekst, a na końcu kreski końcowe?

Tu przydadzą się **argumenty/parametry** funkcji. \

Pamiętaj: kiedy masz przepis funkcji, ona nie pyta się, jaki dokładnie tekst chcesz wgrać. Mówisz jej, że ona ma jakiś tekst, który chcesz przerobić i mówisz, co z tym tekstem ma zrobić.

````python
def dodaj_kreski(tekst):
    print(30 * "-")
    print(tekst)
    print(30 * "-")

dodaj_kreski("Litwo Ojczyzno Moja")
dodaj_kreski("Ty jesteś jak zdrowie")
dodaj_kreski("Ile cię trzeba cenić")
dodaj_kreski("Ten tylko się dowie")
dodaj_kreski("Kto Cię stracił.")
````

### Można też dodawać wiele argumentów:


````python
def dodaj_kreski(tekst, przerywnik):
    print(30 * przerywnik[0]) # Przypominam, że napis to jest lista znaków. Dlatego biorę zawsze pierwszy element tej "listy" (tekstu).
    print(tekst)
    print(30 * przerywnik[0])

dodaj_kreski("Litwo Ojczyzno Moja", "-")
dodaj_kreski("Ty jesteś jak zdrowie", "_")
dodaj_kreski("Ile cię trzeba cenić", "+")
dodaj_kreski("Ten tylko się dowie", "H")
dodaj_kreski("Kto Cię stracił.", "Duży napis")
````
