# Przed IF-em:
### Zmienna logiczna. To zmienna przyjmująca dwie wartości: prawda lub fałsz.

```python
zmienna1 = True
zmienna2 = False

print(zmienna1)
print(zmienna2)
```

### Wyrażenie logiczne

Wyrażenie, które zwraca prawdę lub fałsz:

```python
print(3<2)
print(10 >= 20)

# UWAGA!!!
# a == b oznacza porównanie dwóch zmiennych: a, b;
# a =  boznacza przypisanie zmiennej a wartości b;
print(20 == 10)
print(10 == 10)
```

### Działania logiczne:
**a or b** - zwraca prawdę, jeśli co najmniej jeden element jest prawdziwy
\

**a and b** - zwraca prawdę jeśli a i b są prawdziwe jednocześnie.

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)

print()

print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
print(20 < 18 or 3 == 3)
```

# IF

Przydatny, kiedy chcemy wykonać kod tylko przy spełnionym określonym warunku

```python
# Prosty przykład

wiek = int(input("Ile masz lat? "))

if wiek >= 18:
  print("Gratuluję, jesteś dorosły")

```

```python
# Prosty przykład

wiek = int(input("Ile masz lat? "))


if wiek >= 18:
  print("Gratuluję, jesteś dorosły")
else:
  print("Jesteś dzieckiem")
```
