# Pętla While

### Co, jesli potrzebujemy pętli, która nie będzie miała np. przeglądać danych po liście albo wykonywać sie z góry określoną ilość razy, tylko np. będzie miała działać, dopóki użytkownik nie wpisze niepoprawnej rzeczy?

Składnia wygląda tak: 

```
while warunek_logiczny:
    rob_cos()
```



````python
# Ten program będzie dodawał do siebie liczby wybrane przez użytkownika dopóki ich suma jest mniejsza od 100.
suma = 0

while suma < 100:
    suma += float(input(suma))
    print(f"Obecna suma wynosi {suma}.")

print()
print(f"OStatecznie Twoja suma wynosi {suma}.")
````

### Polecenie **break**
**break** sprawia, że program przerywa działanie całej pętli. \
Teraz program sprawi, że jak wpiszemy napis "Koniec", to wtedy program przerwie działanie pętli.

````python
suma = 0

while suma < 100:
    liczba = input(suma)
    if liczba == "Koniec":
        print("Kończymy.")
        break

    suma += float(liczba)

    print(f"Obecna suma wynosi {suma}.")

print()
print(f"OStatecznie Twoja suma wynosi {suma}.")
````

### Polecenie **continue**.
To polecenie przerywa aktualną iterację (powtórzenie) pętli, ale kontynuuje pętlę.

````python
suma = 0

while suma < 100:
    liczba = input(suma)
    if liczba == "Koniec":
        print("Kończymy.")
        break
        
    if float(liczba) <= 0:
        print("Ale mógłbyś dodawać liczby dodatnie, ziom...")
        continue
      
    suma += float(liczba)

    print(f"Obecna suma wynosi {suma}.")

print()
print(f"OStatecznie Twoja suma wynosi {suma}.")
````

### Pętla nieskończona.
Czasem cchemy, żeby nasz program nie miał z góry ustalonego końca. W takiej sytuacji można zrobić pętlę nieskończoną.

```
while True:
    dzialaj()
```

````python
# Ten program dodaje do siebie kolejne liczby dodatnie, dopóki nie wpiszemy liczby 0.
suma = 0

while True:
    liczba = input(suma)

    if liczba == "Koniec" or liczba == 0:
        print("Kończymy.")
        break

    if float(liczba) < 0:
        print("Ale mógłbyś dodawać liczby dodatnie, ziom...")
        continue
      
    suma += float(liczba)

    print(f"Obecna suma wynosi {suma}.")

print()
print(f"OStatecznie Twoja suma wynosi {suma}.")
````

### break i continue działają też w pętli for:

````python
# Ten program nie doda liczb ujemnych z liczby.
lista = [1, 3, 5, 10, -2, 11, -11, 21, 32, 11, 5]
suma = 0
for element in lista:
    if element < 0:
        print("Za duża liczba")
        continue
    suma += element
````

### Zgadywanie liczb

````python
import random

min = int(input("Wybierz minimalna liczbę: "))
max = int(input("Wybierz maksymalną liczbę: "))

licznik = 0
liczbaLosowana = random.randint(min, max)

while True:
    liczbaZgadywana = int(input("Spróbuj zgadnąć liczbę: "))

    if liczbaLosowana > liczbaZgadywana:
        print("Wybrałeś za małą liczbę. ")
    elif liczbaLosowana < liczbaZgadywana:
        print("Wybrałeś za dużą liczbę. ")
    else:
        print("Gratuluję, udało się :)")
        break
    licznik += 1

print(f"Zgadywałeś {licznik} razy.")
````
