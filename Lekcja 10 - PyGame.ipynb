# PyGame:
PyGame to biblioteka/moduł języka Python przeznaczony do tworzenia gier komputerowych. Zawiera on narzędzia, które pozwalają na tworzenie okna gry oraz na zarządzanie nią.


Z początku PyGame jest trudny w użytku, ponieważ trzeba wiele bardzo prostych rzeczy wprowadzać od podstaw. Mechanizm

### Wstępny szablon:

```python
# Wgrywanie potrzebnych modułów:
import pygame 
from pygame.locals import *


# Rozpoczyna działanie PyGame
pygame.init()

# Dzięki tym dwóm linijkom mamy stałe 60 klatek na sekundę;
clock = pygame.time.Clock()
fps = 60

# Parametry (rozdzielczość)okna zapisane do zmiennych:
SCREEN_WIDTH = 640
SCREEN_HEIGHT = 480

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption('Gierka')


# Ładowanie obrazków do gry (koniecznie zrób plik o nazwie podanej w cudzysowie i wstaw go w folderze projektu);
hero = pygame.image.load('Hero.png')

run = True
while run:

    # Zegar gry
    clock.tick(fps)

    # Rysowane obiektów na ekranie
    screen.fill((0, 0, 30))
    screen.blit(hero, (100, 200)) 


    # To jest TURBOWAŻNE I NIE USUWAJ TEGO!!!
    # Ten kod sprawia, że po kliknięciu krzyżyka w prawym górnym rogu okienko się zamknie.
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False


    # To tworzy nową klatkę gry; :)
    pygame.display.update()

pygame.quit()
```

### Rozwinięcie kodu z zajęć: (wprowadzenie):
**Kilka fukcjonalności zostało dodanych:**
- ruch za pomocą strzałek;
- ruch oparty na fizyce (zmienne playerVelocityX, playerVelocityY);
- dodanie siły grawitacji (ciągłe zwiększanie playerVelocityY);
- dodanie siły oporów (mnożenie prędkości przez 0.98);
- zabezpieczenie obiektu przed wyjściem z planszy;

```python
import pygame
from pygame.locals import *


# Rozpoczyna działanie PyGame
pygame.init()

# Dzięki tym dwóm linijkom mamy stałe 60 klatek na sekundę;
clock = pygame.time.Clock()
fps = 60

# Wymiary okna gry przechowywane w postaci dwóch zmiennych:
SCREEN_WIDTH = 1280
SCREEN_HEIGHT = 720

# screen - Okienko gry (oraz wybranie rozdzielczości ekranu;
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
# Nazwa okienka
pygame.display.set_caption('Piatek 17:30')



# ZMIENNE GRACZA:
# # Zmienne przechowujące pozycje gracza:
playerX = 100
playerY = 200

# # Zmienne przechowujące prędkość gracza:
playerVelocityX = 0
playerVelocityY = 0
# Wczytanie obrazka do obiektu
hero = pygame.image.load('Hero.png')

run = True
while run:

    # Zegar gry - pilnuje, żeby gra działała w 60-ciu FPS-ach.
    clock.tick(fps)

    # LOGIKA GRY:

    # Przechwytywanie klawiszy gry:
    keys = pygame.key.get_pressed()
    if keys[K_UP]:
        playerVelocityY -= 1
    if keys[K_DOWN]:
        playerVelocityY += 1
    if keys[K_LEFT]:
        playerVelocityX -= 1
    if keys[K_RIGHT]:
        playerVelocityX += 1

    # Dodanie siły grawitacji do gry (zwiększa prędkość w każdej klatce);
    playerVelocityY += 0.5

    # Dodanie siły oporu (po prostu hamuje obiekt. Im większa prędkość, tym bardziej hamuje):
    playerVelocityX *= 0.98
    playerVelocityY *= 0.98

  # Kod, który na podstawie zmiennych prędkości nadaje pozycje obiektowi.
    playerX += playerVelocityX
    playerY += playerVelocityY



    # Sprawdzanie, czy gracz nie wyszedł poza ekran;
    if playerX < 0:
        playerX = 0
        playerVelocityX *= -1
    if playerX > SCREEN_WIDTH - hero.get_width():
        playerX = SCREEN_WIDTH - hero.get_width()
        playerVelocityX *= -1

    if playerY < 0:
        playerY = 0
        playerVelocityY = 0
    if playerY > SCREEN_HEIGHT - hero.get_height():
        playerY = SCREEN_HEIGHT - hero.get_height()
        playerVelocityY = 0


    # RYSOWANIE OBIEKTÓW W GRZE:

    # Wypełnienie ekranu gry:
    screen.fill((0, 0, 30))

    # Rysowane obiektów na ekranie
    screen.blit(hero, (playerX, playerY))


    # To jest TURBOWAŻNE I NIE USUWAJ TEGO!!!
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False


    # To tworzy nową klatkę gry;
    pygame.display.update()

pygame.quit()
```
