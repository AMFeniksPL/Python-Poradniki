# Lekcja 12 - Programowanie obiektowe V2

## Podstawy programowania obiektowego:

Tworzenie klas polega na tym, że definiuje własny rodzaj obiektów, tzw. **Klasy**, któremu nadaje się pewne właściwości, mianowicie: Pola i Metody:
- Pola - zmienne przechowywane w klasie;
- Metody - funkcje przypisane do klasy;

Tak jak np. mieliśmy obiekt typu Turtle(), który mogliśmy zapisać do zmiennej albo po prostu zmienne typu int, str, etc...
Tak samo mozemy stworzyć własny typ "zmiennej".

### Klasa a Obiekt:
**Klasą** jest np. Człowiek, który może mieć imię, nazwisko, wzrost, etc...
\

**Obiektem** jest np. Przykładowy Polak, który ma na imię Jan, na nazwisko Nowak i 176cm wzrostu;



## Przykładowy kod tworzący klasę:

```python
class uczen:                                      # Tworzenie Klasy
  def __init__(self, imie, nazwisko, oceny):      # __init__() to tzw. konstruktor, czyli funkcja mówiąca, jak tworzymy nowy obiekt
    self.imie = imie                              # self to informacja, że obiekt odwołuje się do samego siebie - tutaj tworzy sobie zmienną imie;
    self.nazwisko = nazwisko
    self.oceny = oceny                            # Ten konstruktor pozwala nam na wpisanie ucznia z użyciem trzech parametrów



  def wyswietl(self):                             # Tak tworzymy funkcje dodatkowe - tutaj chcę po prostu wyświetlić dane ucznia;
    print("\n---------------------------------------------")
    print(f"Imię: {self.imie};")
    print(f"Nazwisko: {self.nazwisko};")
    print(f"Oceny: {self.oceny};")
    print("---------------------------------------------\n\n\n")



  def srednia_ocen(self):
    return sum(self.oceny)/len(self.oceny)        # Pamiętaj, że jeśli chcesz, żeby program zwracał jakąś wartość, którą można przetworzyć w innej funkcji: musi być return.



  def dodaj_ocene(self, nowa_ocena):
    self.oceny.append(nowa_ocena)
```

### Najwazniejsze informacje:
- Żeby zdefiniować klasę, trzeba użyć formułki `class nazwa_obiektu:`
- żeby stworzyć konstruktor, tzn. sposób na stworzenie nowego obiektu, stwórz funkcję `def __init__(self, parametr1, parametr2):`
- żeby stworzyć zmienne należące do konkretnego obiektu danej klasy, stwórz ją,, pisząc `self.nazwa_zmiennej = wartość`
- żeby stworzyć dodatkową funkcję w klasie (tzn. metodę) użyj formułki `def nazwa_funkcji(self, parametr1, parametr2)`
- pamiętaj, że jeśli chcesz sie gdzieś w kodzie klasy odnieść do jego pola lub metody: poprzedź jej nazwę słówkiem `self.nazwa_elementu`.

## Tworzenie obiektu:

```python
# Tworzenie nowego obiektu - popatrz, że nie używamy czterech parametrów, self wskazuje na sam obiekt;
nowicjusz = uczen("Anna", "Shirley", [2,4,5,3,2,1,2,6,6,4,5])

# Użycie funkcji:
nowicjusz.wyswietl()

# Jeśli funkcja nie ma w sobie printa, to do jej wyświetlenia w konsoli potrzeba słowa print;
print(nowicjusz.srednia_ocen())

nowicjusz.dodaj_ocene(1)
nowicjusz.dodaj_ocene(1)
nowicjusz.dodaj_ocene(1)
nowicjusz.dodaj_ocene(1)
nowicjusz.dodaj_ocene(1)
nowicjusz.wyswietl()
print(nowicjusz.srednia_ocen())

# Wypisanie zmiennej
print(nowicjusz.imie)
```

Stwórzmy klasy: Motorbike (Motocykl) oraz Car (Samochód)

Obie klasy mają następujące rzeczy:
- pola: name, vMax, acceleration, gearCount;
- metody: __init__(), time_to_vMax(), gearUp();

```python
class Motorbike:

    def __init__(self, name, vMax, acceleration, gearCount):
        self.name = name
        self.vMax = vMax
        self.acceleration = acceleration
        self.gearCount = gearCount


        # Te zmienne tworzymy, ale nie przypisujemy im dowolnej wartości, tylko wpisujemy na sztywno;
        self.actualGear = 0
        self.wheelCount = 2

    
    def time_to_vMax(self):
        return self.vMax/self.acceleration

    def gearUp(self):
        if self.actualGear < self.gearCount:
            self.actualGear += 1

class Car:

    def __init__(self, name, vMax, acceleration, gearCount):
        self.name = name
        self.vMax = vMax
        self.acceleration = acceleration
        self.gearCount = gearCount

        self.actualGear = 0
        self.wheelCount = 4


    def time_to_vMax(self):
        return self.vMax/self.acceleration

    def gearUp(self):
        if self.actualGear < self.gearCount:
            self.actualGear += 1
```

```python
yamaha = Motorbike("Yamaha", 300, 50, 6)
maluch = Car("Fiat 126P", 70, 10, 4)

print(yamaha.acceleration)
print(Car.time_to_vMax)
```

### Generalnie Kod jest super, ale... on się trochę powtarza... 

CO JA MOWIĘ: TYLKO ZMIENNA wheelCount JEST INNA.

A co jakby stworzyć klasę, która przechowuje elementy wspólne w jednym miejscu, a we właściwych klasach mamy tylko te unikalne rzeczy?

Otóż da się. I to się nazywa "Dziedziczenie".

Jak to zrobić?
1. Stwórz klasę, która zawiera wspólne informacje;
2. Stwórz nowe klasy, które zawierają unikalne informacje i zrób dziedziczenie od klasy wyżej.

```python
class Vehicle:
    def __init__(self, name, vMax, acceleration, gearCount):
        self.name = name
        self.vMax = vMax
        self.acceleration = acceleration
        self.gearCount = gearCount

        self.actualGear = 1

    def time_to_vMax(self):
        return self.vMax/self.acceleration

    def gearUp(self):
        if self.actualGear < self.gearCount:
            self.actualGear += 1

# W taki sposób można wskazać, że klasa Motorbike dziedziczy od klasy Vehicle;
class Motorbike(Vehicle):

    def __init__(self, name, vMax, acceleration, gearCount):
        # super() daje informacje do klasy nadrzędnej;
        super().__init__(name, vMax, acceleration, gearCount)

        # Wpisujemy tu tylko nowe rzeczy w tej klasie;
        self.wheelCount = 2

    
    # Normalnie tu byśmy wpisywali dodatkowe metody, ale mamy je już w klasie Vehicle - więc są niepotrzebne;


class Car(Vehicle):

    def __init__(self, name, vMax, acceleration, gearCount):
        super().__init__(name, vMax, acceleration, gearCount)
        self.wheelCount = 4
```

```python
# Okazuje się, że klasy Motorbike i Car mogą używać pól i metod, których nie mają w kodzie.
# Dlaczego? Ponieważ mają to wszystko w klasie nadrzędnej Vehicle

yamaha = Motorbike("Yamaha", 300, 50, 6)
maluch = Car("Fiat 126P", 70, 10, 4)

print(yamaha.acceleration)
print(Car.time_to_vMax)
```

Jak widzisz: kodu jest znacznie mniej, a co więcej: jeśli zamarzy nam się, że np. pierwszym biegiem w zmiennej actualGear ma być 0 (luz), a nie 1, to można łatwo to zmienić dla wszystkich reprezentantów.

## Zmienne Statyczne:

Co jakbyśmy chcieli przechowywać jakąś informację wspólną dla wszystkich obiektów, np. ilość stworzonych obiektów? 

Od tego są tzw. zmienne statyczne, tzn. takie, które nie są przypisane do konkretnego obiektu, ale są przypisane do samej klasy. Jak to działa?

```python
class Vehicle:
    # Tworzymy zmienną statyczną. Poza funkcją __init__(), bez self i przypisujemy jej początkową wartość.
    vehicleCount = 0 

    def __init__(self, name, vMax, acceleration, gearCount):
        self.name = name
        self.vMax = vMax
        self.acceleration = acceleration
        self.gearCount = gearCount

        self.actualGear = 1

        # W tym miejscu mówię, że zmienna vehicleCount ma się zwiększyć za każdym razem, jak stworzymy nowy obiekt.
        # Zwróć uwagę, że muszę powiedzieć, że to jest zmienna z klasy Vehicle;
        Vehicle.vehicleCount += 1

    def time_to_vMax(self):
        return self.vMax/self.acceleration

    def gearUp(self):
        if self.actualGear < self.gearCount:
            self.actualGear += 1

class Motorbike(Vehicle):

    def __init__(self, name, vMax, acceleration, gearCount):
        # super() daje informacje do klasy nadrzędnej;
        super().__init__(name, vMax, acceleration, gearCount)

        # Wpisujemy tu tylko nowe rzeczy w tej klasie;
        self.wheelCount = 2

class Car(Vehicle):

    def __init__(self, name, vMax, acceleration, gearCount):
        super().__init__(name, vMax, acceleration, gearCount)
        self.wheelCount = 4
```

```python
print(Vehicle.vehicleCount)
yamaha = Motorbike("Yamaha", 300, 50, 6)
print(Vehicle.vehicleCount)
komar = Motorbike("Komar", 150, 30, 4)
print(Vehicle.vehicleCount)


print("\n\nNumber of vehicles: ")
print(Vehicle.vehicleCount)
print(Motorbike.vehicleCount)
print(Car.vehicleCount)
print(yamaha.vehicleCount)

# Główna zaleta zmiennej statycznej - nie trzeba jej odpalać na konketnym obiekcie, wystarczy wpisać nazwę klasy;
# Można użyć tej zmiennej w każdej klasie oraz na obiekcie.
```

## Klasa abstrakcyjna

Wszystko fajnie, ale teraz takie ważne pytanie: czy nie powinniśy mieć zakazu tworzenia obiektów klasy Vehicle? No bo jak mamy mieć jakąś konkretną liczbę kół, to trudno stworzyć pojazd z niezdefiniowaną liczbą kół, co to: TRANSFORMERS?!

Tu z pomocą przychodzi klasa abstrakcyjna. To po prostu klasa, w której dostęp do niektórych rzeczy mamy tylko z poziomu klasy dziedzicznej.

Przykładowo: tutaj zrobiłem tak, żeby konstruktor klasy Vehicle był metodą abstrakcyjną.

Co to oznacza?

Dokładnie to oznacza tyle, że możesz wykorzystać kod klasy abstrakcyjnej, żeby móc go rozwinąć w klasie dziedzicznej. **Ale nie możesz stworzyć obiektu klasy abstrakcyjnej.**




```python
# Należy zaimportować mechanizmy: ABC oraz abstractmethod;
# Następnie: 

from abc import ABC, abstractmethod
class Vehicle(ABC):
    # Tworzymy zmienną statyczną. Poza funkcją __init__(), bez self i przypisujemy jej początkową wartość.
    vehicleCount = 0 

    @abstractmethod
    def __init__(self, name, vMax, acceleration, gearCount):
        self.name = name
        self.vMax = vMax
        self.acceleration = acceleration
        self.gearCount = gearCount

        self.actualGear = 1

        # W tym miejscu mówię, że zmienna vehicleCount ma się zwiększyć za każdym razem, jak stworzymy nowy obiekt.
        # Zwróć uwagę, że muszę powiedzieć, że to jest zmienna z klasy Vehicle;
        Vehicle.vehicleCount += 1

    def time_to_vMax(self):
        return self.vMax/self.acceleration

    def gearUp(self):
        if self.actualGear < self.gearCount:
            self.actualGear += 1

class Motorbike(Vehicle):

    def __init__(self, name, vMax, acceleration, gearCount):
        # super() daje informacje do klasy nadrzędnej;
        super().__init__(name, vMax, acceleration, gearCount)

        # Wpisujemy tu tylko nowe rzeczy w tej klasie;
        self.wheelCount = 2

class Car(Vehicle):

    def __init__(self, name, vMax, acceleration, gearCount):
        super().__init__(name, vMax, acceleration, gearCount)
        self.wheelCount = 4
```

```python
# Ten kod nie zadziała. bo konstruktor klasy Vehicle jest abstrakcyjny
car1 = Vehicle("asd", 1, 2, 3)
```
